```python
import pandas as pd
import numpy as np
from datetime import date, datetime
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('input\/COVID-19_geo_timeseries_ver_0311.csv', parse_dates=True)
df = df.set_index('update_time')
```

```python
df = df.drop(['data_source', 'country_code', 'region', 'country_flag', 'country'], axis=1)
```

```python
final_df = pd.DataFrame(index = df.index.unique())
for val in df.province.unique():
    df_by_province = df.loc[(df.province == val)]
    df_by_province.drop_duplicates(inplace=True)
    df_by_province = df_by_province.drop(['province'], axis=1)
    df_by_province = df_by_province.add_prefix(str(val)+'_')
    final_df = pd.concat([final_df, df_by_province], axis=1, join='outer', sort=True) 
```
