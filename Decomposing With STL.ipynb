```python
from imputationLibrary import decomposeStl, nature
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from sklearn import preprocessing
```

```python
INPUT = 'output/\CompleteWeeklyIndexes'
```

```python
df = pd.read_csv(INPUT+'.csv', parse_dates=True, index_col=0)
df_filled = df.fillna(0) #+ 1 #Adding 1 is necessary to allow for multiplicative decomposition
```

```python
df_nan = df.isna()
```

```python
def plot_ac(df, name):
    normalized = preprocessing.normalize([np.array(df.fillna(0))])
    corr = signal.correlate(normalized, normalized, mode='full')
    plt.plot(corr[0], 'o-', markersize=2)
    plt.axhline(2/np.sqrt(len(df)), ls=':')
    plt.axhline(-2/np.sqrt(len(df)), ls=':')
    plt.title(name + ' auto-correlation')
    plt.show()
```

```python
for i in range(0,len(df_filled.columns)):
    df_decomposed = decomposeStl.decomposeStl(df_filled.iloc[:,i])
    df_decomposed.plot()
    plt.show()
    print("Resid is white noise? ", nature.isWhiteNoise(df_decomposed.resid))
    print("Resid is seasonal noise? ", nature.isSeasonal(df_decomposed.resid))
    print("Resid is trended noise? ", nature.isTrended(df_decomposed.resid))
    print("Resid is seasonal and trended noise? ", nature.isTrendedAndSeasonal(df_decomposed.resid))
    plot_ac(df_decomposed.resid, "Resid")
    plot_ac(df_decomposed.trend, "Trend")
    plot_ac(df_decomposed.seasonal, "Seasonal")
```
